```python
class PDController:
    def __init__(self, kp, kd):
        self.kp = kp  # Assuming scalar values for a single joint
        self.kd = kd  # Assuming scalar values for a single joint
        self.previous_error = 0.0  # Scalar, since we're controlling one joint at a time

    def compute_control(self, current_position, target_position, dt):
        # Calculate the error as a scalar
        error = target_position - current_position
        
        # Calculate the derivative of the error as a scalar
        d_error = (error - self.previous_error) / dt
        
        # Update the previous error for the next iteration
        self.previous_error = error
        
        # Calculate the control action (torque) for the joint as a scalar
        torque = self.kp * error + self.kd * d_error
        
        return torque
```

```python
def tune_joint(controller, joint_index, target_position, num_steps=1000, dt=0.01):
    """
    Tune a single joint by adjusting Kp and Kd, observing its response to a target position.
    
    :param controller: An instance of PDController pre-initialized with starting Kp, Kd values.
    :param joint_index: Index of the joint to be tuned.
    :param target_position: The target position for the joint.
    :param num_steps: Number of simulation steps for the tuning process.
    :param dt: Time step for the simulation.
    """
    # Arrays to store data for plotting
    time_steps = np.arange(num_steps) * dt
    target_positions = np.full(num_steps, target_position)
    actual_positions = []
    
    for step in range(num_steps):
        # env.render()
        
        current_position = env.current_sim_state.legs_qpos[joint_index]  # Get current position of the joint
        torque = controller.compute_control(current_position, target_position, dt)
        
        # Apply torque only to the joint being tuned
        action = np.zeros(12)  # Assuming 12 joints
        action[joint_index] = torque
        
        # Step the environment with the action
        env.step(action)
        
        # Record the actual position for plotting
        actual_positions.append(env.current_sim_state.legs_qpos[joint_index])
    
    # # Plot the results
    # plt.figure(figsize=(10, 6))
    # plt.plot(time_steps, target_positions, label='Target Position', linestyle='--', color='red')
    # plt.plot(time_steps, actual_positions, label='Actual Position', marker='', color='blue')
    # plt.xlabel('Time (s)')
    # plt.ylabel('Position (rad)')
    # plt.title(f'Tuning Joint {joint_index}')
    # plt.legend()
    # plt.grid(True)
    # plt.show()

    # # Return the recorded positions for further analysis if needed
    # return actual_positions

    # Plotting section
    plt.figure(figsize=(10, 6))
    plt.plot(time_steps, target_positions, label='Target Position', linestyle='--', color='red')
    plt.plot(time_steps, actual_positions, label='Actual Position', marker='', color='blue')
    plt.xlabel('Time (s)')
    plt.ylabel('Position (rad)')
    plt.title(f'Tuning Joint {joint_index}')
    plt.legend()
    plt.grid(True)
    
    # Save the plot to a file
    pltname = "plot" + str(joint_index)
    plt.savefig(pltname)
    plt.close()  # Close the plot to free up memory

    sys.exit()
    
    # Optional: Print a message to indicate saving
    print(f"Plot saved to" + "pltname")
```
